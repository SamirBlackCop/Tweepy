```python
from itertools import count
import pandas as pd

import matplotlib.pyplot as plt

from matplotlib.animation import FuncAnimation
%matplotlib notebook
#%matplotlib inline
```

```python
plt.style.use('fivethirtyeight')
```

```python
frame_len = 10000
fig = plt.figure(figsize=(7,4))

def animate(i):
    data = pd.read_csv('sentiment.csv')
    y1= data['Trump']
    y2= data['Biden']
    
    if len(y1) <=frame_len:
        plt.cla()                          
        plt.plot(y1, label="Donald Trump")
        plt.plot(y2, label="Biden")
        
    else:                                               # This is optional 
        plt.cla()
        plt.plot(y1[-frame_len], label="Donald Trump")
        plt.plot(y2[-frame_len], label="Biden")
        
    plt.legend(loc='upper left')
    plt.tight_layout()                                  # Gives autometic padding to layout
    
    
an1 = FuncAnimation(plt.gcf(), animate, interval=100)   # neede paremeter




# interval parameter use for how often we run the code 1000 means 1 sec
# plt.gcf() use to get current figures
# plt.gca() use to get current axes
# plt.cla() cleares the axes so each interval of time we get same line colur for same axes or graph
# frame length use for how many time want to open the graph window as example when 10000 data are updated then close this func
```
